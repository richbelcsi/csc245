```python
import pandas as pd
data = pd.read_csv('company_sales_data.csv')
data
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(data['month_number'], data['total_profit'], marker='o')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Company profit per month')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['month_number'], data['total_profit'], linestyle='--', color='red', marker='o', markerfacecolor='black', linewidth=3, label='Profit data of last year')
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.title('Company Sales data of last year')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

for product in data.columns[1:7]:  
    plt.plot(data['month_number'], data[product], marker='o', label=product)
    
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('Sales data')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['month_number'], data['toothpaste'], color='blue', label='Toothpaste Sales')
plt.xlabel('Month Number')
plt.ylabel('Number of units sold')
plt.title('Toothpaste Sales Data')
plt.grid(True, linestyle='--')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(data['month_number'] - 0.2, data['facecream'], width=0.4, color='blue', label='Face Cream Sales Data')
plt.bar(data['month_number'] + 0.2, data['facewash'], width=0.4, color='orange', label='Face Wash Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('Facewash and Facecream Sales Data')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(data['month_number'], data['bathingsoap'], color='blue', label='Bathing Soap Sales')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('Bathing Soap Sales Data')
plt.grid(True)
plt.savefig('bathing_soap_sales.png')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['total_profit'], bins=10, color='#007ac3')
plt.xlabel('Profit Range In Dollars')
plt.ylabel('Actual Profit In Dollars')
plt.title('Profit Data')
plt.grid(False)
plt.show()
```

```python
last_year_data = data[data['month_number'] >= data['month_number'].max() - 11]

total_sales = last_year_data.iloc[:, 1:7].sum()

plt.figure(figsize=(8, 8))
plt.pie(total_sales, labels=total_sales.index, autopct='%1.1f%%', colors=['#007ac3', 'orange', 'green', 'red', 'purple', 'brown'], startangle=0)
plt.title('Sales Data')
plt.legend(loc='lower right')
plt.show()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

axs[0].plot(data['month_number'], data['bathingsoap'], color='black', marker='o', label='Bathing Soap Sales')
axs[0].set_xlabel('Month Number')
axs[0].set_ylabel('Sales units in number')
axs[0].set_title('Bathing Soap Sales Data')
axs[0].grid(False)

axs[1].plot(data['month_number'], data['facewash'], color='red', marker='o', label='Facewash Sales')
axs[1].set_xlabel('Month Number')
axs[1].set_ylabel('Sales units in number')
axs[1].set_title('Facewash Sales Data')
axs[1].grid(False)

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.stackplot(data['month_number'], data['facecream'], data['facewash'], data['toothpaste'],
              data['bathingsoap'], data['shampoo'], data['moisturizer'],
              labels=['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer'])
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Product Sales Data per Month (Stack Plot)')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
