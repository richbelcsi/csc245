```python
df = pd.read_csv('titanic.csv')

def age_group(age):
    if age < 18:
        return 'child'
    else:
        return 'adult'

df['ageGroup'] = df['age'].apply(age_group)

pivot_table = df.pivot_table(index=['sex', 'ageGroup'], columns='pclass', values='survived', aggfunc='mean')

print("Survival Rate by Gender, Age Group, and Class:")
print(pivot_table)
```

```python
age_bins = [0, 10, 30, 60, 80]
age_labels = ['0-10', '10-30', '30-60', '60-80']

df['agecategory'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)

print("First few rows with AgeCategory:")
print(df[['age', 'agecategory']].head())
```

```python
age_bins = [0, 10, 30, 60, 80]
age_labels = ['0-10', '10-30', '30-60', '60-80']

df['agecategory'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)

pivot_table = df.pivot_table(index=['sex', 'agecategory'], columns='pclass', values='survived', aggfunc='sum', fill_value=0)

print("Survival Count by Gender, Age Category, and Class:")
print(pivot_table)
```

```python
age_bins = [0, 10, 30, 60, 80]
age_labels = ['0-10', '10-30', '30-60', '60-80']

df['agecategory'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)

pivot_table = df.pivot_table(index=['sex', 'agecategory'], columns='pclass', values='survived', aggfunc='mean')

print("Survival Rate by Gender, Age Category, and Class:")
print(pivot_table)
```

```python
age_bins = [0, 10, 30, 60, 80]
age_labels = ['0-10', '10-30', '30-60', '60-80']

df['agecategory'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)

fare_bins = [0, df['fare'].median(), df['fare'].max()]
fare_labels = ['low fare', 'high fare']
df['farecategory'] = pd.cut(df['fare'], bins=fare_bins, labels=fare_labels, right=False)

pivot_table = df.pivot_table(index=['sex', 'agecategory'], columns=['farecategory', 'pclass'], values='survived', aggfunc='mean')

print("Survival Rate by Gender, Age Category, Fare Category, and Class:")
print(pivot_table)
```
